.step 1 Remove Missing Values

```python
data[data == ""] = "unknown@gmail.com"
```

2. Remove Duplicate Rows

```python
# Convert to DataFrame to use drop_duplicates, then back to numpy array
df_temp = pd.DataFrame(data)
df_unique_temp = df_temp.drop_duplicates()
data = df_unique_temp.values
```

3. Standardize Text

```python
# Uppercase names
data[:,1] = np.char.upper(data[:,1].astype(str))

# Lowercase emails
data[:,2] = np.char.lower(data[:,2].astype(str))

# Capitalize city
data[:,3] = np.char.title(data[:,3].astype(str))
```

step 4 Remove Invalid Values

```python
quantity = data[:,6].astype(int)

# Keep only positive quantity
data = data[quantity > 0]
```

DATA PREPARATION
 Convert Data Types

```python
quantity = data[:,6].astype(int)
price = data[:,7].astype(int)
```

6. Create New Column (Total Amount)

```python
total = quantity * price

# Add new column
data = np.column_stack((data, total))
```

```python
print(data)
```

data visualizations

 Bar Chart

```python
import matplotlib.pyplot as plt

products = data[:,4]
prices = data[:,7].astype(int)

plt.bar(products, prices)
plt.xlabel("Product")
plt.ylabel("Price")
plt.title("Product vs Price")
plt.show()
```

Line Chart

```python
plt.plot(prices)
plt.xlabel("Index")
plt.ylabel("Price")
plt.title("Price Trend")
plt.show()
```

5. Pie Chart (Product Distribution)

```python
unique_products, counts = np.unique(products, return_counts=True)

plt.pie(counts, labels=unique_products, autopct='%1.1f%%')
plt.title("Product Distribution")
plt.show()
```

6. Histogram (Price Distribution)

```python
plt.hist(prices)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()
```
